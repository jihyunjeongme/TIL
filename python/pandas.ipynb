### Reindex

```python
s2 = s.reindex(
    ['a', 'c', 'e', 'g']
)
s2
```

```python
# Do not use this
s1 = pd.Series([0, 1, 2], index=[0, 1, 2])
s2 = pd.Series([3, 4, 5], index=['0', '1', '2'])
s1
s2
```

```python
s1 + s2
```

```python
# How to fix? Method 1
s1 = pd.Series([0, 1, 2], index=[0, 1, 2])
s2 = pd.Series([3, 4, 5], index=['0', '1', '2'])
```

```python
s2.index = s2.index.astype(int)
```

```python
s2
```

```python
s2.index
```

```python
s1 + s2
```

```python
# How to fix it? Method 2
s1 = pd.Series([0, 1, 2], index=[0, 1, 2])
s2 = pd.Series([3, 4, 5], index=['0', '1', '2'])
```

```python
s1.index = ['a', 'b', 'c']
s2.index = ['a', 'b', 'c']
```

```python
s1 + s2
```

#### `reindex()` useful Arguments

- `fill_value`

```python
s2 = s.copy()
s2
```

```python
s2.reindex(['a', 'f'])
```

```python
s2.reindex(['a', 'f'], fill_value=0)  # fill 0 insteand of Nan
```

- `method`

```python
s3 = pd.Series(['red', 'green', 'blue'], index=[0, 3, 5])
s3
```

```python
s3.reindex(np.arange(0,7))
```

```python
# based on previous one, fill in NaN
s3.reindex(np.arange(0,7), method='ffill')
```
